# Titanic Dataset - Exploratory Data Analysis (EDA)

### Import Libraries

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# For better visuals
sns.set(style="whitegrid")
plt.style.use("ggplot")
```

### Load Dataset

```python
# Load Titanic dataset from seaborn
titanic_data = sns.load_dataset('titanic')

# Preview the data
titanic_data.head()
```

### Basic Data Info

```python
# Dataset shape
print("Shape:", titanic_data.shape)

# Info and data types
titanic_data.info()

# Numerical summary
titanic_data.describe()

# Categorical summary
titanic_data.describe(include='object')
```

### Missing Values Analysis

```python
# Missing value count
missing = titanic_data.isnull().sum()
print("Missing Values:\n", missing[missing > 0])

# Visualize missing values
plt.figure(figsize=(10, 5))
sns.heatmap(titanic_data.isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

### Univariate Analysis

#### Categorical Columns

```python
categorical_cols = ['sex', 'class', 'embarked', 'who', 'deck', 'alive', 'alone']

for col in categorical_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(data=titanic_data, x=col, palette="pastel")
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

#### Numerical Columns

```python
numeric_cols = titanic_data.select_dtypes(include=np.number).columns

titanic_data[numeric_cols].hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

### Outlier Detection with Boxplots

```python
for col in numeric_cols:
    plt.figure(figsize=(5, 3))
    sns.boxplot(data=titanic_data, x=col, color='lightblue')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

### Bivariate Analysis
#### Survival by Category

```python
sns.barplot(data=titanic_data, x='sex', y='survived')
plt.title("Survival Rate by Sex")
plt.show()

sns.barplot(data=titanic_data, x='class', y='survived', order=['First', 'Second', 'Third'])
plt.title("Survival Rate by Class")
plt.show()
```

#### Age vs Survival

```python
sns.kdeplot(data=titanic_data[titanic_data['survived']==1], x='age', label='Survived', fill=True)
sns.kdeplot(data=titanic_data[titanic_data['survived']==0], x='age', label='Did Not Survive', fill=True)
plt.title("Age Distribution by Survival")
plt.legend()
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(titanic_data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
