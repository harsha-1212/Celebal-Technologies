```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('tip.csv')
data.head(5)
```

```python
x = data['day']
y = data['total_bill']

plt.bar(x, y)

plt.title("Bar chart")

plt.ylabel('Total Bill')

plt.xlabel('Day')

plt.show()
```

```python
x = data['total_bill']

plt.hist(x)

plt.title("Histogram")

plt.ylabel('Frequency')

plt.xlabel('Total Bill')

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={ 'fontsize': 18})

plt.plot(gas.Year, gas.USA, 'r.-.', label='United States')
plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'b.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-.', label='Australia')


plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```
